```python
import numpy as np
import torch
import torch.nn as nn #package for neural network
import pandas as pd
from sklearn.preprocessing import StandardScaler
from torch.utils.data import Dataset
```

```python
#load the data
data=pd.read_csv('diabetes.csv')
data
```

```python
#extract input =x ie all rows and all cols except the last which class
#convert the data to  numpy array using the iloc method
x=data.iloc[:,0:-1].values
y_string =list(data.iloc[:,-1])
```

```python
y_string
```

```python
x.shape
```

```python
#convert the y string to intergers
y_int =[]
for ch in y_string:
  if ch =='positive':
    y_int.append(1)
  else:
    y_int.append(0)
```

```python
#convert list to numpy array and convert to float 64 since x is float also
y = np.array(y_int,dtype='float64')
y
```

x' = x-u/std

```python
#Feature normalization,all features should have the same range(-1,1)
sc = StandardScaler()
x = sc.fit_transform(x)
print(x)
```

```python
#convert numpy arrays to pytorch tensors
x=torch.tensor(x)
y=torch.tensor(y)
print(x.shape)
print(y.shape)
```

```python
#add dimensionality to the y data since we will BCE loss
y=y.unsqueeze(1)
print(y.shape)
print(y)
```

```python
#load data to your dataloader for batch procesing and shuffling
class Dataset(Dataset):
  def __init__(self,x,y):
    self.x =x
    self.y=y
  def __getitem__(self, index):
    return self.x[index],self.y[index]
  def __len__(self):
    return len(self.x)
```

```python
dataset = Dataset(x,y)
len(dataset)
```

```python
train_loader=torch.utils.data.DataLoader(dataset,
                            batch_size=32,
                            shuffle=True)
```

```python
train_loader
```

```python
print("There is {} batches in the dataset".format(len(train_loader)))
for (x,y) in train_loader:
  print("For one iteration/batch , there is:")
  print(f"Data:{x.shape}")

  print(f"labels: {y.shape}")
  break
```

```python
#build the neural network
#7 inputs features ->
class Model(nn.Module):
  def __init__(self,input_features,output_features):
    super(Model,self).__init__()
    self.fc1 = nn.Linear(input_features,5)
    self.fc2 = nn.Linear(5,4)
    self.fc3 = nn.Linear(4,3)
    self.fc4 = nn.Linear(3,output_features)
    self.sigmoid = nn.Sigmoid()
    self.tanh = nn.Tanh()
  def forward(self,x):
    out = self.fc1(x)
    out = self.tanh(out)
    out = self.fc2(out)
    out = self.tanh(out)
    out = self.fc3(out)
    out = self.fc4(out)
    out = self.sigmoid(out)
    return out
```

```python
#create object of our model
net = Model(7,1)
#size average =True --> the losses are averaged over observations for each mini batch
criterion=torch.nn.BCELoss(size_average=True)
#we will use the SGD with momentum with lr 0.1
optimizer  =torch.optim.SGD(net.parameters(),lr=0.1,momentum=0.9)
```

```python
#Train the neural network
epochs = 200
for epoch in range(200):
  for inputs,labels in train_loader:
      inputs = inputs.float()
      labels = labels.float()
        # Forward propagation
      outputs = net(inputs) #
      #calculate loss
      loss=criterion(outputs,labels)
      # clear gradient buffer to avoid accumulation of gradient (w = w-lr*gradient)
      optimizer.zero_grad()
      #backpropagation
      loss.backward()
      #update weight
      optimizer.step()
  # calculate accuracy for one epoch
  # 1. threshold the ouptut if it is bigger than 0.5 we get 1 if smaller than 0.5=> 0
  output = (outputs>0.5).float()
  # calculate mean
  accuracy = (output == labels).float().mean()
  #print statics
  print("Epoch {}/{},Loss:{:.3f},Accuracy:{:.3f}".format(epoch+1,epochs,loss,accuracy))


```
